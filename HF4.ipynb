```
# The code was tested in Collab.

# Installing required modules
!pip install openimages
from openimages.download import download_images

# Downloading the images
base_dir = '/HF4/openimages'
download_images(base_dir, ["Airplane", "Cake", "Bird"], csv_dir=base_dir, limit=600, exclusions_path=None)

category1 = 'airplane'
category2 = 'cake'
category3 = 'bird'
```

```
# Saving the images to directories, so that the "flow_from_directory" function will accept it
import os
# Before sorting
download_dir_1 = os.path.join(base_dir, category1 + '/images')
download_dir_2 = os.path.join(base_dir, category2 + '/images')
download_dir_3 = os.path.join(base_dir, category3 + '/images')
fnames_1 = os.listdir(download_dir_1)
fnames_2 = os.listdir(download_dir_2)
fnames_3 = os.listdir(download_dir_3)

train_fnames_1 = fnames_1[:400]
train_fnames_2 = fnames_2[:400]
train_fnames_3 = fnames_3[:400]

valid_fnames_1 = fnames_1[400:500]
valid_fnames_2 = fnames_2[400:500]
valid_fnames_3 = fnames_3[400:500]

test_fnames_1 = fnames_1[500:600]
test_fnames_2 = fnames_2[500:600]
test_fnames_3 = fnames_3[500:600]

from shutil import copy, rmtree





new_base_dir = '/HF4/data'
train_dir = os.path.join(new_base_dir, 'Train')
valid_dir = os.path.join(new_base_dir, 'Valid')
test_dir = os.path.join(new_base_dir, 'Test')
train_dir_1 = os.path.join(train_dir, category1)
train_dir_2 = os.path.join(train_dir, category2)
train_dir_3 = os.path.join(train_dir, category3)

valid_dir_1 = os.path.join(valid_dir, category1)
valid_dir_2 = os.path.join(valid_dir, category2)
valid_dir_3 = os.path.join(valid_dir, category3)

test_dir_1 = os.path.join(test_dir, 'unlabeled')

# Deleting folder: 'data'
try:
  rmtree(new_base_dir)
except:
  True

# Creating the necessary folders
# https://medium.com/@vijayabhaskar96/tutorial-image-classification-with-keras-flow-from-directory-and-generators-95f75ebe5720
os.mkdir(new_base_dir)
os.mkdir(train_dir)
os.mkdir(valid_dir)
os.mkdir(test_dir)
os.mkdir(train_dir_1)
os.mkdir(train_dir_2)
os.mkdir(train_dir_3)
os.mkdir(valid_dir_1)
os.mkdir(valid_dir_2)
os.mkdir(valid_dir_3)
os.mkdir(test_dir_1)


# Copying the files
for fname in train_fnames_1:
  copy(os.path.join(download_dir_1, fname), train_dir_1)
for fname in train_fnames_2:
  copy(os.path.join(download_dir_2, fname), train_dir_2)
for fname in train_fnames_3:
  copy(os.path.join(download_dir_3, fname), train_dir_3)


for fname in valid_fnames_1:
  copy(os.path.join(download_dir_1, fname), valid_dir_1)
for fname in valid_fnames_2:
  copy(os.path.join(download_dir_2, fname), valid_dir_2)
for fname in valid_fnames_3:
  copy(os.path.join(download_dir_3, fname), valid_dir_3)

for fname in test_fnames_1:
  copy(os.path.join(download_dir_1, fname), test_dir_1)
for fname in test_fnames_2:
  copy(os.path.join(download_dir_2, fname), test_dir_1)
for fname in test_fnames_3:
  copy(os.path.join(download_dir_3, fname), test_dir_1)



```

```
# Let's plot some of these pictures
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Number of rows & number of columns
nrows = 4
ncols = 4

tmp_1 = [os.path.join(download_dir_1, fname) for fname in fnames_1[:int(nrows)]]
tmp_2 = [os.path.join(download_dir_2, fname) for fname in fnames_2[:int(nrows)]]
tmp_3 = [os.path.join(download_dir_3, fname) for fname in fnames_3[:int(nrows)]]

plt.figure(figsize=(15, 5 * nrows))
for i, img_path in enumerate(tmp_1 + tmp_2 + tmp_3):
  sub_plot = plt.subplot(nrows, ncols, i + 1)
  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```

```
# Downloading & training pretrained networks (transfer learning)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras import backend as K
import numpy as np

img_height = 299
img_width = 299

# Loading pretrained model - without the fully connected layers -
base_model = InceptionV3(weights = 'imagenet', include_top = False)

#Then, we add new layers to the end of the pretrained model (transfer learning)
# We use global average pooling after the last convolutional layer so as to flatten the 2D convolution
x = base_model.output
x = GlobalAveragePooling2D()(x)
# Adding fully connected layer
x = Dense(1024, activation = 'relu')(x)

# Output layer
predictions = Dense(3, activation = 'softmax')(x)
# Creating the model
model = Model(inputs = base_model.input, outputs = predictions)

```

```
# We will train only the last layers we have added
# We freeze the weights of the pretrained network & only train the konvolutional layers
for layer in base_model.layers:
  layer.trainable = False

# Compile model
# Because we have multiple classes, we use categorical crossentropy loss
model.compile(optimizer='adam', metrics=['accuracy'],loss='categorical_crossentropy')

# Preparing the picture for loading and data augmentation
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(train_dir, target_size=(img_height, img_width), batch_size=20, class_mode='categorical', classes=[category1, category2, category3])
validation_generator = test_datagen.flow_from_directory(valid_dir, target_size=(img_height, img_width), batch_size=20, class_mode='categorical', classes=[category1, category2, category3])
```

```
# The summary of the model
# model.summary()
```

```
# The following function trains & augments the data at the same time
history = model.fit_generator(train_generator,steps_per_epoch=10,validation_data=validation_generator,validation_steps=10,epochs=5)
```

```
model.save('HF4.h5')

# Plotting the training and validation accuracy loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))

import matplotlib.pyplot as plt
plt.figure()
plt.plot(epochs, acc, 'bo', label = 'Training accuracy')
plt.plot(epochs, val_acc, 'b', label = 'Validation accuracy')
plt.legend()

plt.figure()
plt.plot(epochs, loss, 'ro', label = 'Training loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation loss')
plt.legend()
```

```
# Then, let's train the upper layers
for layer in model.layers[:172]:
       layer.trainable = False
for layer in model.layers[172:]:
       layer.trainable = True

# We need to recompile
model.compile(optimizer=SGD(lr=0.0001, momentum=0.9), metrics=['accuracy'], loss='binary_crossentropy')
history2 = model.fit_generator(train_generator,steps_per_epoch=10,validation_data=validation_generator,validation_steps=10,epochs=3)
print("End of learning.")
```

```
model.save('HF4_2.h5')

# Plotting the training and validation accuracy loss
acc = history2.history['accuracy']
val_acc = history2.history['val_accuracy']
loss = history2.history['loss']
val_loss = history2.history['val_loss']
epochs = range(len(acc))

import matplotlib.pyplot as plt
plt.figure()
plt.plot(epochs, acc, 'bo', label = 'Training accuracy')
plt.plot(epochs, val_acc, 'b', label = 'Validation accuracy')
plt.legend()

plt.figure()
plt.plot(epochs, loss, 'ro', label = 'Training loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation loss')
plt.legend()
```

```
# Here we look at some examples for predictions
# The label tells us what the network predicted for the image below.

from tensorflow.keras.preprocessing.image import img_to_array, load_img

img_categories = ["Airplane", "Cake", "Bird"]

# Number of rows & number of columns
nrows = 4
ncols = 4

tmp_1 = [os.path.join(test_dir_1, fname) for fname in test_fnames_1[:int(nrows)]]
tmp_2 = [os.path.join(test_dir_1, fname) for fname in test_fnames_2[:int(nrows)]]
tmp_3 = [os.path.join(test_dir_1, fname) for fname in test_fnames_3[:int(nrows)]]

plt.figure(figsize=(15, 5 * nrows))

# Plotting the images & adding predicted labels
for i, img_path in enumerate(tmp_1 + tmp_2 + tmp_3):
  sub_plot = plt.subplot(nrows, ncols, i + 1)
  img = mpimg.imread(img_path)
  plt.imshow(img)

  img = load_img(img_path, target_size=(img_height, img_width))  # this is a PIL image
  x = img_to_array(img) * (1. / 255)  # Numpy array with shape (img_height, img_width, 3)
  x = x.reshape((1,) + x.shape)  # Numpy array with shape (1, img_height, img_width, 3)
  prediction = model.predict(x)
  plt.title(img_categories[np.argmax(prediction[0])])

plt.show()
```
